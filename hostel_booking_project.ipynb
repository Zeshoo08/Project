```python



import pandas as pd
import numpy as np


df = pd.read_csv("hotels_dataset.csv")

# Display First 5 Rows
print("First 5 Rows:\n")
print(df.head())



# Question 1
# Function for Data Information
# ============================================

def data_info(dataframe):

    result = pd.DataFrame({
        "Data Type": dataframe.dtypes,
        "Null Values": dataframe.isnull().sum(),
        "Null Percentage": (dataframe.isnull().sum() / len(dataframe)) * 100,
        "Unique Values": dataframe.nunique()
    })

    return result

print("\n==============================")
print("Question 1 Result")
print("==============================\n")

print(data_info(df))



# Question 2
# Reservation Cancellation Count


print("\n")
print("Question 2 Result")
print("\n")

cancel_data = df['is_canceled'].value_counts()

print(cancel_data)

# Percentages
cancel_percentage = (cancel_data / len(df)) * 100

print("\nCancellation Percentages:\n")
print(cancel_percentage)

# Conclusion
print("\nConclusion:")

print(f"Visitors who canceled reservation: {cancel_percentage[1]:.2f}%")
print(f"Visitors who did NOT cancel reservation: {cancel_percentage[0]:.2f}%")



# Question 3
# Hotel-wise Cancellation Percentage


print("\n==============================")
print("Question 3 Result")
print("==============================\n")

# City Hotel
city_hotel = df[df['hotel'] == 'City Hotel']

city_cancel_percentage = city_hotel['is_canceled'].mean() * 100

print(f"a) City Hotel Cancellation Percentage: {city_cancel_percentage:.2f}%")


# Resort Hotel
resort_hotel = df[df['hotel'] == 'Resort Hotel']

resort_cancel_percentage = resort_hotel['is_canceled'].mean() * 100

print(f"b) Resort Hotel Cancellation Percentage: {resort_cancel_percentage:.2f}%")


# Compare
print("\nc) Conclusion:")

if city_cancel_percentage > resort_cancel_percentage:
    print("City Hotel has the bigger percentage of canceled reservations.")
else:
    print("Resort Hotel has the bigger percentage of canceled reservations.")

# Question 4
# Visitors who did not cancel reservation

print("\n")
print("Question 4 Result")
print("\n")

df_checkout = df[df['is_canceled'] == 0]

print(df_checkout.head())

print("\nShape of df_checkout:")
print(df_checkout.shape)
```
